# Competency Profiles - Inter Reater Agreement

-----

The example scores in this notebook stem from a german questionnaire on Physic's Teachers Pedagogicaln Content Knowledge.

## Setup

```python
import qutools.data as qtdata
import qutools.scoring as qtscrng
```

```python
quconfig = qtdata.QuConfig.from_yaml("quconfigs/physics-pck.yaml")

quird = qtscrng.QuInterraterAnalysis(
    quconfig=quconfig,
    df_rater1="<NOT IN THIS REPO>/pck-scores-coder1.csv",
    df_rater2="<NOT IN THIS REPO>/pck-scores-coder2.csv",
)
```

```python
quird.evaluate()
```

```python
_ = quird.plot_confusion_matrix()
```
